### Multivariate Analysis

#### Question 9: Perform a multivariate analysis to explore relationships between the important variables in the dataset. (It is a good idea to explore relations between numerical variables as well as relations between numerical and categorical variables) [15 marks]

```python
# plotting a pairplot to check for the relationship between all the variables

sns.pairplot(df, hue="Make", diag_kind="kde")
plt.show()
```

* People of age less than 30 prefer Hatchback. Those of age between 25 and 45 prefer Sedan. SUVs have been bought only by those aged between 35 and 60 as their salary is also high.
* People with number of dependents 0, 3 and 4 have not bought a sedan.
* In general, Hatchback cars cost the least. Sedans cost more than Hatchbacks. SUVs cost the highest.
* People with total salary more than apprx. 125000 have purchased only SUVs.

```python
# plotting a boxplot to check how the distribution of salary for each Make of the car

plt.figure(figsize=(10,4))
sns.boxplot(data = df, x = "Make", y = "Total_salary", hue = "Make")
plt.show()
```

* Customers with high salary have prefered SUVs as they cost more.
* Customers with salary more than 100000 dollars mostly prefer SUVs.

```python
# plotting a boxplot to check the distribution of the Price for different professions.

plt.figure(figsize=(10,4))
sns.boxplot(data = df, x = "Profession", y = "Price", hue = "Profession")
plt.show()
```

* Salaried people bought costly cars as compared to business people.


```python
# plotting a bar graph to check the distribution of price for people with different marital status

plt.figure(figsize=(10,4))
sns.boxplot(data = df, x = "Marital_status", y = "Price", hue = "Profession")
plt.show()
```

* Among both married and single people, Salaried people have bought costlier cars.
* Especially among the singles, business people have mostly purchased lower priced cars.

```python
# boxplot to check the distribution of price for different Make of cars for different professions

plt.figure(figsize=(8, 5))
sns.boxplot(x="Make", y="Price", hue="Profession", data=df)
plt.title("Car Price by Car Type & Profession")
plt.show()
```

* The distribution of the price of the car by the car type and profession is not heavily skewed for any categories.
* The distribution is similar for Business people and salaried people for each of the car type.

```python
# plotting a bar chart to check the preferance of car type for people with different educational background

car_counts = df.groupby(["Make", "Education"])["Make"].count().unstack()
car_counts.plot(kind="bar", stacked=True, figsize=(8, 5))
plt.title("Car Type Preference by Education")
plt.ylabel("Number of Customers")
plt.show()
```

* Post graduates prefer Hatchbacks considerably more than SUVs and sedans.
* SUVs are preferred the least by both graduates and post graduates. This might be because of the high pricing of the SUVs.


```python
# plotting a count plot to check the distribution of the types of car by personal loan availed or not

sns.countplot(data=df, x='Personal_loan', hue='Make');
```

* People who have and who have not availed a personal loan have made similar kind of purchase for different kinds of cars but those who have availed a personal have purchased very little number of SUVs.

```python
sns.countplot(data=df, x='House_loan', hue='Make');
```

```python
# plotting a correlation matrix for Age, Price and Total salary

col_list = ['Age', 'Total_salary', 'Price']
sns.heatmap(df[col_list].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

* Positive correlation of 0.79 between price and age shows that higher the age, higher the price of the cars bought.
* A positive correlation of 0.47 between salary and age shows that salary tends to increase as age increases.
* There is a positive correlation between total salary and price: 0.35, but it is not as high as we would expect.
